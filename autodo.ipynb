```python
import torch.nn.functional as F
import torch.nn as nn

s = torch.Tensor([[1],[2]])

weights = torch.tensor([[0.5, 0.5, 0.5],
                        [0.5, 0.5, 0.5],
                        [0.5, 0.5, 0.5]])
nb_channels = 3

x = torch.ones(2, nb_channels, 7, 7)
a, b, c=x.shape
print(b)

weights_list = []
for _ in range(2):
  weight = nn.Parameter(s[_]*weights.unsqueeze(0).unsqueeze(0).repeat(3,1,1,1))
  weights_list.append(weight)

outputs=[]
for idx in range(2):
  input = x[idx:idx+1]
  filt = weights_list[idx]
  output = F.conv2d(input, filt, groups=3, padding='same')
  outputs.append(output)

outputs = torch.stack(outputs)
outputs = outputs.squeeze(1)
print(outputs)

```

```python
import numpy as np
import scipy.stats as st
import torch

def gkern(kernlen=21, nsig=3):
    """Returns a 2D Gaussian kernel."""

    x = np.linspace(-nsig, nsig, kernlen+1)
    kern1d = np.diff(st.norm.cdf(x))
    kern2d = np.outer(kern1d, kern1d)
    return torch.Tensor(kern2d/kern2d.sum())

print(gkern(5,1))
```

```python
def ikern(kernlen=5):
  x = torch.zeros([kernlen, kernlen])
  x[int((kernlen - 1)/2), int((kernlen - 1)/2)] = 1
  return x

print(ikern(7))
```
